## The principle of intermediates

First of all, why do we need to think about file formats at all, as long as it is relatively easy to read them? Because the effort in developing a workflow scales poorly with the number of tasks and the number of formats if we address the problem naively. I also talk more about this principle on [my blog](http://negfeedback.blogspot.com/2012/11/the-pattern-of-intermediate-forms-and.html)

I'll use the excellent [Networkx](http://networkx.github.io/) library to produce some graphs to prove my point.

```python
import networkx
```

Let's say there are $N$ formats and $M$ final products.

```python
N = 10
M = 10
```

```python
sources = ['S{}'.format(i) for i in range(N)]
targets = ['T{}'.format(i) for i in range(M)]
```

```python
# Make a graph showing the connectivity between sources and targets
connectivity = networkx.Graph([(source, target) for source in sources for target in targets])
# color the sources red and the targets blue
color_lookup = {'S': 'red', 'T': 'blue'}
node_colors = [color_lookup[node[0]] for node in connectivity.nodes()]
```

Notice how many connections there are!

```python
networkx.draw_circular(connectivity, node_color=node_colors)
```

There are $M\times N$ routes between the sources and targets. That's a lot of code.

```python
len(connectivity.edges())
```

This problem is alleviated by introducing an intermediate format

```python
color_lookup['i'] = 'yellow'
withintermediates = networkx.Graph([(source, 'intermediate') for source in sources] 
                                    + [('intermediate', target) for target in targets])
node_colors = [color_lookup[node[0]] for node in withintermediates.nodes()]
```

```python
networkx.draw(withintermediates, node_color=node_colors)
```

```python
len(withintermediates.edges())
```

So we've gone from $N\times M$ to $N+M$. That's pretty good going. The win is even bigger when we consider that many of the "backend" translators (starting from the intermediate and going to the target) can be written and maintained by someone else.

## Basic text files

CSV files are just text files. Problem solved, right? Let's see how to combined data is handled. This is the output from a TGA (thermogravimetric analysis) device.

We don't even need Pandas to read basic text files, right?

```python
lines = open('samplefiles/tga.txt').readlines()
lines[0]
```

Oops, I thought they said it was a text file!

## Diversion: Unicode
You really need to read [this article](http://www.joelonsoftware.com/articles/Unicode.html)

The if you look closely at the output above, it seems like every second character is useful. This is a clue that they may be using UTF-16.

```python
lines[0][2::2]
```

We could throw away half the data, or we could decode it properly. The `io` module provides an encoding-aware wrapper around this for Python 2.7 and up. In Python 3 this is included in the base language.

```python
import io
lines = io.open('samplefiles/tga.txt', encoding='utf-16').readlines()
lines[0]
```

This file contains a tabular part as well as other metadata.

I've written a basic parser for these files. It separates out the information in the file into header information and the tabular data part. I recommend that you open the text file in an editor to get a feeling for what it looks like, then peruse my parser

```python
import tga
tga = tga.tga('samplefiles/tga.txt')
```

```python
tga.header
```

```python
tga.data.head()
```

## The emerging standard: JSON

The kind of information stored in the header is well modelled as a nested dictionary structure. This kind of data is well handled by [JSON](http://www.json.org). Lots of people use JSON, and parsers are already built into many languages.

Unfortunately doesn't open in Excel by default

```python
import json
json.dump(tga.header, open('samplefiles/header.json', 'w'))
```

```python
!cat samplefiles/header.json
```

```python
json.load(open('samplefiles/header.json'))
```

So we could store the header in an JSON file and the tabular data in a CSV. Isn't there a way to keep this information together?

## What we should be doing: HDF5

Yes. The [HDF5 standard](http://www.hdfgroup.org) is the complete solution to all the data problems I've discussed here. Unfortunately industry has not enmbraced it, partly, I think because it doesn't open in Excel.

```python
import h5py
import numpy
```

```python
f = h5py.File('samplefiles/tga.hdf5', 'w')
```

```python
dataset = f.create_dataset('tga_analysis', data=tga.data)
```

```python
for key, value in tga.header.iteritems():
    dataset.attrs[key] = unicode(value)
```

```python
f.close()
```

You can open this new file in HDFview to see the hierarchy and produce simple plots.
